## Central LImit Theorem

The Central Limit Theorem states that the means of sufficiently large samples taken from any population will be distributed in a normal fashion around the population mean

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

First, we will check whether the central limit theorem applies to variables that follow a normal distribution. 


```python
x=np.random.normal(size=1000)
```

How normal distribution graph look
</br>A normal distribution graph, also known as a bell curve, has a symmetrical shape with a single peak at the center. The x-axis represents the values of the data being measured, while the y-axis represents the frequency of occurrence of each value. The curve is characterized by the mean and standard deviation of the data, with the majority of the data falling within one or two standard deviations of the mean. The tails of the curve extend indefinitely in both directions, indicating that extreme values are possible but rare

```python
sns.distplot(x)
plt.show()
```

To prove the Central Limit Theorem, we need to compute the mean of large samples taken from the population set and then graph it to determine if the theorem holds true or not.

```python
def sample_mean_calculator(population,sample_size,num_samples):
    sample_means = []
    for i in range(num_samples):
        sample=np.random.choice(population,size=sample_size,replace=False)
        sample_mean=sample.mean()
        sample_means.append(sample_mean)
    return sample_means
```

We can check the applicability of the Central Limit Theorem on a population set that follows a normal distribution


```python
sns.distplot(x)
plt.show()
```

Checking Central limit theorem

```python
sns.distplot(sample_mean_calculator(x,5,5))
plt.show()
```

```python
sns.distplot(sample_mean_calculator(x,10,5))
plt.show()
```

```python
sns.distplot(sample_mean_calculator(x,15,10))
plt.show()
```

```python
sns.distplot(sample_mean_calculator(x,25,20))
plt.show()
```

```python
sns.distplot(sample_mean_calculator(x,30,20))
plt.show()
```

Larger is the sample size more it will tighter towards population mean

```python
sns.distplot(sample_mean_calculator(x,50,20))
plt.show()
```

```python
sns.distplot(sample_mean_calculator(x,1000,1000))
plt.show()
```
